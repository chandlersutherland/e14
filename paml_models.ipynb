First step, see which model fits the gene the best. Then, identify number of sites under selection 

```python
#write a function that reads an output file, and saves the lines of each lnL for the model, then the line ranges for positively selected sites. 
def number_hunter(paml_path):
    with open(paml_path, 'r') as file: 
        lnl_index=[]
        BEB_index=[]
        codon_len=np.nan
        #file_contents=file.readlines()
        #codon_len=file_contents[0].split('   ')[2].strip('\n')
        for l_no, line in enumerate(file):
            if l_no == 0: 
                line_string=line.split(' ')
                while("" in line_string):
                    line_string.remove("")
                codon_len=int(line_string[1].strip('\n'))
            if 'lnL' in line:
                #print('lnL found at: ', l_no)
                lnl_index.append(l_no)
            if 'BEB' in line:
                #print('BEB begins at: ', l_no)
                BEB_index.append(l_no)
            if 'The grid' in line:
                #print('BEB ends at:', l_no)
                BEB_index.append(l_no)
        if np.isnan(codon_len):
            print('no output paml file')
    return lnl_index, BEB_index, codon_len
```

lnL is the log of the likelihood of the model. We want the LRT statistic, which is twice the difference in the log-likelihood between the null and alternative hypotheses, 2(l1-l0), where l0 is the log-likelihood of the null, and l1 is the log-likelihood of the alternative. 

The LRT statistic is then compared with the X2 distribution with DOF equal to the number of free parameters.

```python
#next, write a function that takes in the bounds of the BEB analysis and returns a dataframe 
#paml returns all positions with a posterior prob > 0.5 
def file_converter(start, end, file_path): 
    #takes in coords found by number hunter of BEB info, reads and cleans, then returns the df. 
    with open(file_path, 'r') as file: 
        positions=[]
        file_contents=file.readlines()
        for l_no, line in enumerate(file_contents[start+6:end-3]):
            line_string=line.split(' ')
            while("" in line_string):
                line_string.remove("")
            #print(line_string)
            position=int(line_string[0])
            aa=line_string[1]
            post_prob=float(line_string[2].strip('*').strip('*'))
            post_mean=float(line_string[3])
            se=float(line_string[5].strip('\n'))
            row_res={'position':position, 'aa':aa, 'post_prob':post_prob, 'post_mean':post_mean, 'se':se}
            positions.append(row_res)
        df=pd.DataFrame.from_records(positions, columns=['position', 'aa', 'post_prob', 'post_mean', 'se'])
    return df
```

```python
#finally, a function that brings it all together 
#want to return a row that has clade, the lrt results, and # positions with posterior prob > 95 
#tried to write to be able to skip files that don't have results 
def wrapper(i):
    file_path=paml_logs[i]
    clade=file_path.split('/')[-2]
    index=number_hunter(file_path)
    if index[0] == []:
        print('Could not process '+clade+'. Check.')
        row_ret={'clade':clade, 'm1_m2_lrt':np.nan, 'm7_m8_lrt':np.nan, 'm2_count_95':np.nan, 'm8_count':np.nan, 'codon_len':np.nan}
    if index[0] != []:
        m1_m2=model_comp(index[0][0], index[0][1], file_path)
        m7_m8=model_comp(index[0][2], index[0][3], file_path)

        positions_m2=file_converter(index[1][0], index[1][1], file_path)
        positions_m8=file_converter(index[1][2], index[1][3], file_path)
        
        #write out positions files if I ever want them again 
        output_dir=file_path.strip(clade+'_pos_paml.mlc')
        positions_m2.to_csv(output_dir+'model2_positions.csv')
        positions_m8.to_csv(output_dir+'model8_positions.csv')

        m2_count=sum(positions_m2['post_prob']>0.95)
        m8_count=sum(positions_m8['post_prob']>0.95)
        codons=index[2]

        row_ret={'clade':clade, 'm1_m2_lrt':m1_m2, 'm7_m8_lrt':m7_m8, 'm2_count_95':m2_count, 'm8_count':m8_count, 'codon_len':codons}
    return row_ret 
```

```python
#put it all together 
paml_output=[]
for i in ready_pamls:
    #print(i)
    row=wrapper(i)

    paml_output.append(row)

df=pd.DataFrame.from_records(paml_output)
df
```
